```python
import pandas as pd
from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import train_test_split

melbourne_file_path = '../Melborn House Price Prediction/melb_data.csv'
melbourne_data = pd.read_csv(melbourne_file_path)
melbourne_data.describe()
```

```python
melbourne_data.columns
```

```python
y = melbourne_data.Price
features = ['Rooms', 'Landsize', 'Bathroom', 'Lattitude', 'Longtitude']
X = melbourne_data[features]
X.head(n=50)
```

```python
train_X, val_X, train_y, val_y = train_test_split(X, y, random_state=1)
```

```python
#Randomforest model
from sklearn.ensemble import RandomForestRegressor
rf_model = RandomForestRegressor()
rf_model.fit(train_X, train_y)
rf_val_predictions = rf_model.predict(val_X)
rf_val_mae = mean_absolute_error(rf_val_predictions, val_y)

print("Validation MAE for Random Forest Model: {:,.0f}".format(rf_val_mae))
```

```python
#LinearRegression model
from sklearn.linear_model import LinearRegression
lr_model = LinearRegression()
lr_model.fit(train_X, train_y)
lr_val_predictions = lr_model.predict(val_X)
lr_val_mae = mean_absolute_error(lr_val_predictions, val_y)

print("Validation MAE for LinearRegressiont Model: {:,.0f}".format(lr_val_mae))
```

```python
def get_mae(max_leaf_nodes, train_X, val_X, train_y, val_y):
    lr_model = LinearRegression()
    lr_model.fit(train_X, train_y)
    lr_val_predictions = lr_model.predict(val_X)
    lr_val_mae = mean_absolute_error(val_y, lr_val_predictions)
    return(lr_val_mae)

for max_leaf_nodes in [5, 50, 500, 5000]:
    my_mae = get_mae(max_leaf_nodes, train_X, val_X, train_y, val_y)
    print("Max leaf nodes: %d  \t\t Mean Absolute Error:  %d" %(max_leaf_nodes, my_mae))
```

Linear Regression is not a tree-based model. It works by finding the best-fitting linear relationship between the features and the target variable. For Linear Regression, there isn't a direct equivalent hyperparameter like max_leaf_nodes to tune its complexity in the same way tree-based models are tuned. Linear Regression's complexity is primarily determined by the number of features in your dataset.
